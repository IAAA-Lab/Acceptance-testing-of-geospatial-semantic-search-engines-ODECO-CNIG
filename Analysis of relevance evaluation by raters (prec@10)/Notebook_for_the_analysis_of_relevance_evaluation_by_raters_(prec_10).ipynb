```python
# Import libraries
import pandas as pd
from pyirr import read_data, kappam_fleiss
```

```python
# Read file containing relevance scores by three raters for 5 queries in 4 information retrieval systems
data=pd.read_csv("Relevance evaluation.csv", sep=";")
data
```

```python
# Calculation of prec@10 scores by query and system
evaluation_rater=data.pivot_table(index = ["system","query","result"], 
              columns = "rater" , 
              values = "relevance", 
              aggfunc = "sum").reset_index()

evaluation_rater["sum_relevance"]=evaluation_rater["Rater 1"]+evaluation_rater["Rater 2"]+evaluation_rater["Rater 3"]
evaluation_rater.loc[evaluation_rater["sum_relevance"] >= 2, "final_relevance"] = 1
evaluation_rater.loc[evaluation_rater["sum_relevance"] < 2, "final_relevance"] = 0

evaluation_rater.pivot_table(index = ["query"], 
              columns = "system" , 
              values = "final_relevance", 
              aggfunc = "mean").reset_index()
```

```python
# Calculation of Fleiss' Kappa for the 3 raters
kappam_fleiss(evaluation_rater[["Rater 1","Rater 2","Rater 3"]].to_numpy(), detail=True)
```
